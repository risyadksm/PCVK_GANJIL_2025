```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')
```

```python
img = np.zeros(shape=(500, 650, 3), dtype=np.uint8)

cv2.line(img, pt1=(20, 40), pt2=(580, 40), color=(255, 255, 255), thickness=2)

cv2.rectangle(img, pt1=(40, 80), pt2=(260, 200), color=(0, 255, 255), thickness=3)

cv2.rectangle(img, pt1=(300, 80), pt2=(560, 200), color=(255, 0, 0), thickness=-1)

cv2.circle(img, center=(140, 330), radius=75, color=(0, 0, 255), thickness=3)

font = cv2.FONT_HERSHEY_DUPLEX
cv2.putText(img, text='OpenCV Shapes', org=(40, 280), fontFace=font, fontScale=1, color=(255, 255, 255), thickness=2, lineType=cv2.LINE_AA)

cv2.arrowedLine(img, pt1=(40, 400), pt2=(250, 400), color=(255, 255, 0), thickness=3)

cv2.circle(img, center=(350, 330), radius=60, color=(0, 255, 0), thickness=-1)

cv2.ellipse(img, center=(530, 370), axes=(70, 40), angle=20, startAngle=0, endAngle=360, color=(255, 255, 0), thickness=3)

pts_pentagon = np.array([[470, 260],
                         [540, 310],
                         [510, 380],
                         [440, 380],
                         [410, 310]], np.int32)
pts_pentagon = pts_pentagon.reshape((-1, 1, 2))

cv2.polylines(img, [pts_pentagon], isClosed=True, color=(255, 0, 255), thickness=3)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Tugas Praktikum D3 - Soal No 1")
plt.axis('on')
plt.show()
```
